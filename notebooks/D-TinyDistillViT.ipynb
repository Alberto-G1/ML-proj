```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.utils.prune as prune
import timm
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from sklearn.metrics import f1_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
from prettytable import PrettyTable
from termcolor import colored
from colored import fg, attr
import seaborn as sns

# Set device (cuda if available, else cpu) without error messages
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load Data with Optimized Transformations
def get_dataloader(batch_size=16):
    transform = transforms.Compose([
        transforms.Resize((128, 128)),  # Resize the images to 128x128
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    ])
    train_dataset = datasets.ImageFolder(root='C:\\Users\\user\\Desktop\\CSC YEAR 2\\SEM 2\\MACHINE LEARNING\\dataset3\\USK-Coffee\\train', transform=transform)
    test_dataset = datasets.ImageFolder(root='C:\\Users\\user\\Desktop\\CSC YEAR 2\\SEM 2\\MACHINE LEARNING\\dataset3\\USK-Coffee\\test', transform=transform)
    val_dataset = datasets.ImageFolder(root='C:\\Users\\user\\Desktop\\CSC YEAR 2\\SEM 2\\MACHINE LEARNING\\dataset3\\USK-Coffee\\val', transform=transform)

    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, test_loader, val_loader

train_loader, test_loader, val_loader = get_dataloader()

class TinyViT(nn.Module):
    def __init__(self, num_classes=10, image_size=128, dim=256, depth=3, heads=4, mlp_dim=512, patch_size=16):
        super(TinyViT, self).__init__()

        self.num_classes = num_classes
        self.patch_embed = nn.Conv2d(3, dim, kernel_size=patch_size, stride=patch_size)

        # Transformer blocks
        self.transformer_blocks = nn.ModuleList([
            nn.TransformerEncoderLayer(d_model=dim, nhead=heads, dim_feedforward=mlp_dim)
            for _ in range(depth)
        ])

        # Pooling & classification layer (Ensuring correct output shape)
        self.pool = nn.AdaptiveAvgPool2d((1, 1))  # Global average pooling
        self.fc = nn.Linear(dim, num_classes)  # Fully connected layer for classification

    def forward(self, x):
        x = self.patch_embed(x)  # Shape: (batch, dim, 8, 8) for 128x128 images with patch size 16

        # Flatten spatial dimensions for transformer input
        x = x.flatten(2).permute(0, 2, 1)  # Shape: (batch, num_patches, dim)

        # Pass through transformer blocks
        for block in self.transformer_blocks:
            x = block(x)

        # Take the mean across patch embeddings
        x = x.mean(dim=1)  # Shape: (batch, dim)

        # Classification head
        x = self.fc(x)  # Shape: (batch, num_classes)

        return x  # Output shape: (batch_size, 10)



# Define Semantic Convolutional Neural Network (SCNN)
# Example forward pass in SemanticCNN or MobileViTWrapper (ensure correct output size)
class SemanticCNN(nn.Module):
    def __init__(self, num_classes=10):
        super(SemanticCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(128 * 16 * 16, 512)  # Adjusted for 128x128 input size
        self.fc2 = nn.Linear(512, num_classes)  # Output layer should have num_classes
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = self.pool(self.relu(self.conv3(x)))
        x = x.view(x.size(0), -1)  # Flatten the output for the fully connected layers
        x = self.relu(self.fc1(x))
        return self.fc2(x)  # Output shape will be [batch_size, num_classes]

# Apply Sparsity
def apply_sparsity(model, amount=0.3):
    for name, module in model.named_modules():
        if isinstance(module, (nn.Linear, nn.Conv2d)):
            prune.l1_unstructured(module, name="weight", amount=amount)
    return model

teacher_model = TinyViT().to(device)
student_model = SemanticCNN().to(device)
apply_sparsity(student_model)

class DistillationLoss(nn.Module):
    def __init__(self, temperature=3.0, alpha=0.7):
        super(DistillationLoss, self).__init__()
        self.temperature = temperature
        self.alpha = alpha
        self.kl_loss = nn.KLDivLoss(reduction="batchmean")
        self.ce_loss = nn.CrossEntropyLoss()

    def forward(self, student_logits, teacher_logits, labels):
        assert student_logits.shape == teacher_logits.shape, f"Shape mismatch: {student_logits.shape} vs {teacher_logits.shape}"
        # Apply softmax to teacher and log_softmax to student
        soft_targets = nn.functional.log_softmax(student_logits / self.temperature, dim=1)
        teacher_targets = nn.functional.softmax(teacher_logits / self.temperature, dim=1)

        # KL loss expects both inputs to have the same shape and distribution
        loss_kl = self.kl_loss(soft_targets, teacher_targets) * (self.temperature ** 2)

        # Cross-entropy loss for student
        loss_ce = self.ce_loss(student_logits, labels)

        return self.alpha * loss_ce + (1 - self.alpha) * loss_kl

# Compute Accuracy & F1 Score
def compute_metrics(model, dataloader):
    model.eval()
    correct, total = 0, 0
    all_preds, all_labels = [], []

    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    accuracy = 100 * correct / total
    f1 = f1_score(all_labels, all_preds, average='weighted')
    return accuracy, f1

# Train Model with Mixed Precision Optimization
def train_model(teacher, student, train_loader, val_loader, test_loader, epochs=10, lr=0.001):
    teacher.eval()
    student.train()
    optimizer = optim.Adam(student.parameters(), lr=lr)
    loss_fn = DistillationLoss()
    scaler = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else torch.amp.GradScaler()  # Handle mixed precision

    train_accs, val_accs, test_accs = [], [], []
    train_f1s, val_f1s, test_f1s = [], [], []

    for epoch in range(epochs):
        total_loss = 0
        correct_train, total_train = 0, 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            with torch.no_grad():
                teacher_outputs = teacher(images)

            with torch.cuda.amp.autocast():  # Mixed precision
                student_outputs = student(images)
                loss = loss_fn(student_outputs, teacher_outputs, labels)

            optimizer.zero_grad()
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            total_loss += loss.item()

            _, predicted = torch.max(student_outputs, 1)
            correct_train += (predicted == labels).sum().item()
            total_train += labels.size(0)

        train_acc = 100 * correct_train / total_train
        val_acc, val_f1 = compute_metrics(student, val_loader)
        test_acc, test_f1 = compute_metrics(student, test_loader)

        train_accs.append(train_acc)
        val_accs.append(val_acc)
        test_accs.append(test_acc)
        train_f1s.append(f1_score([0]*len(train_loader.dataset), [0]*len(train_loader.dataset), average='weighted')) # Placeholder for train F1
        val_f1s.append(val_f1)
        test_f1s.append(test_f1)

        print(f"{fg('blue')}Epoch {epoch+1}/{epochs}{attr('reset')} | Loss: {total_loss/len(train_loader):.4f} | "
              f"{fg('green')}Train Acc: {train_acc:.2f}%{attr('reset')} | "
              f"{fg('yellow')}Val Acc: {val_acc:.2f}%{attr('reset')} | Val F1-score: {val_f1:.4f} | "
              f"{fg('red')}Test Acc: {test_acc:.2f}%{attr('reset')} | Test F1-score: {test_f1:.4f}")

    # Tabulate results
    table = PrettyTable()
    table.field_names = ["Metric", "Train", "Validation", "Test"]
    table.add_row(["Accuracy", f"{train_acc:.2f}%", f"{val_acc:.2f}%", f"{test_acc:.2f}%"])
    table.add_row(["F1-score", "N/A", f"{val_f1:.4f}", f"{test_f1:.4f}"])
    table.add_row(["Learning Rate", lr, lr, lr])
    print(table)

    # Plot training and validation accuracy
    plt.figure(figsize=(10, 6))
    plt.plot(train_accs, label='Training Accuracy', color='green')
    plt.plot(val_accs, label='Validation Accuracy', color='yellow')
    plt.plot(test_accs, label='Test Accuracy', color='red')
    plt.title('Accuracy Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.show()

    # Plot F1 scores
    plt.figure(figsize=(10, 6))
    plt.plot(val_f1s, label='Validation F1-score', color='yellow')
    plt.plot(test_f1s, label='Test F1-score', color='red')
    plt.title('F1-score Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('F1-score')
    plt.legend()
    plt.show()

    # Bar chart for final results
    plt.figure(figsize=(8, 6))
    plt.bar(['Train', 'Validation', 'Test'], [train_acc, val_acc, test_acc], color=['green', 'yellow', 'red'])
    plt.title('Final Accuracy Comparison')
    plt.xlabel('Dataset')
    plt.ylabel('Accuracy (%)')
    plt.show()

    # Compute and display confusion matrix for test set
    student.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = student(images)
            _, predicted = torch.max(outputs, 1)
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Blues')
    plt.xlabel("Predicted labels")
    plt.ylabel("True labels")
    plt.show()

# Train the student model
train_model(teacher_model, student_model, train_loader, val_loader, test_loader)
```
